# 2. INHERITANCE AND TYPE

Pengertian:
* Inheritance : pewarisan atau turunan.
* Inheritance adalah konsep pemrograman dimana sebuah class (induk) dapat ‘menurunkan’ property dan method yang dimilikinya kepada class lain(anak). Class yang akan ‘diturunkan’ bisa disebut sebagai class induk (parent class), super class, atau base class. Sedangkan class yang ‘menerima penurunan’ bisa disebut sebagai class anak (child class), sub class.

Type dari Inheritance :
* a. Pass adalah kondisi dimana suatu method ini diberikan pass karena sama dengan method induk , dan methodnya bisa ditulis ataupun tidak karena tidak berpengaruh.
* b. Overriding adalah kondisi dimana suatu method yang berasal dari class induk ingin anda modifikasi atau ditambahkan beberapa fitur sesuai kebutuhan pada class anak, disinilah peran dari 'overriding method'.
* c. super() adalah mengembalikan atribut dan metode dari super objek (induk) suatu kelas.(mengambil atribut dari induk).
* d. Multiple Inheritance adalah kondisi dimana kita dapat mewariskan dari beberapa level class, superclass dengan mewariskan ke subclass kemudian subclass mewariskan ke class lain.

# 4. ENCAPSULATION

Pengertian:
* Encapsulation adalah cara untuk mengatur komponen class dengan cara menembunyikan alur komponen class tersebut. Komponen class adalah property dan method dari class, kita membuat akses terbatas pada property dan method agar hanya property dan method tertebtu saja yang bisa diakses dengan cara tertentu.

* variabel public (+) : Pendeklarasian variabel tanpa underscore. Variabel public dapat diakses diluar ataupun di dalam class.
* variabel private (-_) : Variabel private tidak dapat diakses di luar class agar dapat diakses maka memakai encapsulation, dengan cara memberi underscore lalu nama class dan nama method. Variabel private hanya dapat diubah dalam method kelas dan tidak di luar kelas.
* underscore_: Variabel private, tidak boleh diakses secara langsung. Tetapi masih bisa diakses secara langsung/konvensi.
* double underscore__: Variabel private, lebih sulit diakses tetapi masih memungkinkan.

# 5. OPERATOR OVERLOADING

* Operator overloading adalah suatu keadaan ketika ada 2 method dalam satu class dengan nama yang sama tetapi dengan argumen yang berbeda, maka yang bisa dipanggil hanya satu method saja, yaitu yg terakhir saja. Dan di python tidak ada method overloading yang ada hanya operator overloading.
* operator overloading : operator dapat membaca addition. dicapai dengan mendefinisikan metode khusus dalam definisi kelas, memungkinkan Anda mendefinisikan kembali makna operator.
* method overloading : method dapat membaca fungsi apa yg dipanggil. 
* python tidak mendukung metode overloading di python namanya special method (magic method : ada double underscore__ di awal dan di akhir).
